```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

model_name = "roberta-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(
    model_name,
    num_labels=2,
    output_attentions=True  # For lookback attention
)

# Financial Test
fin_prompt = "Stock market performed very well today"
fin_input = tokenizer(
    fin_prompt,
    return_tensors="pt",
    padding=True,
    truncation=True
)

with torch.no_grad():
    fin_output = model(**fin_input)

# Extract Attention
attentions = fin_output.attentions
last_layer = attentions[-1]
print("Financial Attention Shape:", last_layer.shape)
print("Financial Logits:", fin_output.logits)

# Medical Test
med_prompt = "Patient diagnosed with severe pneumonia"
med_input = tokenizer(
    med_prompt,
    return_tensors="pt",
    padding=True,
    truncation=True
)

with torch.no_grad():
    med_output = model(**med_input)

print("Medical Logits:", med_output.logits)
print("RoBERTa loaded successfully!")
```
