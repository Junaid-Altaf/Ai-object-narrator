```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

model_name = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(
    model_name,
    num_labels=3,
    output_attentions=True
)

def test_domain(prompt, domain_name):
    inputs = tokenizer(
        prompt,
        return_tensors="pt",
        truncation=True,
        padding=True
    )

    with torch.no_grad():
        outputs = model(**inputs)

    # Extract lookback attention
    attentions = outputs.attentions
    last_attention = attentions[-1]
    lookback = last_attention[0].mean(dim=0)[-1]

    print(f"\n=== {domain_name} Domain ===")
    print(f"Logits: {outputs.logits}")
    print(f"Lookback Attention Shape: {lookback.shape}")
    print(f"Max Attention Score: {lookback.max().item():.4f}")

# Test both domains
test_domain(
    "Revenue increased by 20% this quarter",
    "Financial"
)
test_domain(
    "Patient shows symptoms of respiratory disease",
    "Medical"
)
print("\nDistilBERT completed!")
```
